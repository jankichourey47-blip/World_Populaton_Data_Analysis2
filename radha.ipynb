```python
import requests
from bs4 import BeautifulSoup
```

```python
URL="https://www.scrapethissite.com/pages/simple/"
res=requests.get(URL)
if res.status_code==200:
    print(res.text)
else:
    print("Failed to Fetch Code")
```

```python
with open('data_prime/data.html', 'w',encoding='utf-8') as f:
    f.write(res.text)
```

```python
from bs4 import BeautifulSoup
with open('data_prime/data.html', 'r',encoding='utf-8') as f:
   html_content= f.read()

soup=BeautifulSoup(html_content,"lxml")
all_h3=soup.find_all("h3")
all_countries=[]
for h3 in all_h3:
   name=h3.get_text(strip=True)
   population=h3.find_next("div").select("span.country-population")[0].get_text(strip=True)
   print(h3.find_next("div").select("span.country-population")[0].get_text(strip=True))
   print(h3.attrs)
   capital=h3.find_next("div").select("span.country-capital")[0].get_text(strip=True)
   area=h3.find_next("div").select("span.country-area")[0].get_text(strip=True)
   all_countries.append([name,capital,population,area])
   

```

```python
all_countries
```

```python
import pandas as pd
df=pd.DataFrame(all_countries,columns=["Name","Capital","Population","Area"])
df
```

```python
df.to_csv("cleaned_data/data1.csv",index=False)
```

```python
# DATA HANDLING AND CLEANING USING PYTHON
df.isnull().sum()
df.duplicated()
```

```python
df.columns=["Country_Name","Capital","Population","Area"]
df.columns
# df["Area"].dtype
```

```python
df["Area"]=pd.to_numeric(df["Area"],errors="coerce")
```

```python
df["Capital"]=df["Capital"].fillna("No Capital")
df["Capital"]
```

```python
df["Population"]=df["Population"].astype(int)
df["Population"]
```

```python
df["Population_Density"]=df["Population"]/df["Area"]
df["Population_Density"]
df["Population_Millions"]=df["Population"]/1000000
df["Population_Millions"]
```

```python
def category(pop):
    if pop>100000000:
        return "High"
    elif pop>10000000:
        return"Medium"
    else:
        return"Low"
df["Population_category"]=df["Population"].apply(category)
df["Population_category"]
```

```python
def density_cat(d):
    if d>500: 
        return "Highly-Dense"
    elif d>100:
        return "Moderate"
    else:
        return "Low Density"

df["Density_Category"]=df["Population_Density"].apply(density_cat)
df["Density_Category"]
```
